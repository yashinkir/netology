```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

# Берем номер документа и возвращаем имя владельца, если документ найден. Если не найден, то None
def find_owner_by_document(doc_number):
    for document in documents:
        if document['number'] == doc_number:
            return document['name']
    return None

# возвращает номер полки, на которой хранится документ, или None
def find_shelf_by_document(doc_number):
    for shelf, docs in directories.items():
        if doc_number in docs:
            return shelf
    return None

# Возвращает список с  информацией о всех документах
def list_all_documents():
    result = []
    for document in documents:
        shelf = find_shelf_by_document(document['number'])
        result.append(f"№: {document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {shelf}")
    return result

# Добавляет новую полку, если её еще нет
def add_shelf(shelf_number):
    if shelf_number in directories:
        return False
    directories[shelf_number] = []
    return True

# Удаляет пустую полку, если она пустая
def delete_shelf(shelf_number):
    if shelf_number not in directories:
        return None
    if directories[shelf_number]:
        return False
    del directories[shelf_number]
    return True

# Обрабатываем команды в цикле до тех пор, пока пользователь не введет "q", чтобы выйти из программы
# Бесит большое количество elif...
def main():
    while True:
        command = input("Введите команду:\n")
        if command == "q":
            print("Завершение работы")
            break
        elif command == "p":
            doc_number = input("Введите номер документа:\n")
            owner = find_owner_by_document(doc_number)
            if owner:
                print(f"Владелец документа: {owner}")
            else:
                print("Документ не найден.")
        elif command == "s":
            doc_number = input("Введите номер документа:\n")
            shelf = find_shelf_by_document(doc_number)
            if shelf:
                print(f"Документ хранится на полке: {shelf}")
            else:
                print("Документ не найден.")
        elif command == "l":
            for document_info in list_all_documents():
                print(document_info)
        elif command == "ads":
            shelf_number = input("Введите номер полки:\n")
            if add_shelf(shelf_number):
                print(f"Полка добавлена. Текущий перечень полок: {', '.join(directories.keys())}.")
            else:
                print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories.keys())}.")
        elif command == "ds":
            shelf_number = input("Введите номер полки:\n")
            result = delete_shelf(shelf_number)
            if result is None:
                print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories.keys())}.")
            elif result is False:
                print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {', '.join(directories.keys())}.")
            else:
                print(f"Полка удалена. Текущий перечень полок: {', '.join(directories.keys())}.")
        else:
            print("Неизвестная команда. Попробуйте снова.")


main()
```
